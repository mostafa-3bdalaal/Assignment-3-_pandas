```python
sns.countplot(x='Gender',hue='Gender',data=df)
xlabel='Gender'
ylabel='Count'
title='Gender Distribution'                       #(Gender)
plt.xlabel=(xlabel)
plt.ylabel=(ylabel)
plt.title=(title)
plt.show()
```

```python
df['Payment Method'].value_counts()
```

```python
plt.pie(df['Payment Method'].value_counts(),labels=df['Payment Method'].value_counts().index,autopct='%1.1f%%')
plt.title=('Payment Method Distribution')           #   
plt.show()
```

```python
plt.pie(df['Shipping Type'].value_counts(),labels=df['Shipping Type'].value_counts().index,autopct='%1.1f%%')
plt.title=('Shipping Type Distribution')           #    
plt.show()
```

```python
sns.countplot(x='Color',hue='Color',data=df)
xlabel='Color'
ylabel='Count'
title=('Color Distribution')                       #  
plt.xticks(rotation=70)
plt.show()
```

```python
sns.countplot(x='Size',hue='Size',data=df)
xlabel='Size'
ylabel='Count'                                    # 
title=('Size Distribution')
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
sns.countplot(x='Size',hue='Color',data=df)
plt.xlabel=('Size')
plt.ylabel=('Color')                                #        
plt.title=('Most Common Colors per Size')
plt.legend(loc='center left',bbox_to_anchor=(1,0.5))

plt.show()
```

```python
sns.countplot(x='Season',hue='Season',data=df)
xlabel='Season'
ylabel='Count'                                  # 
title=('Season Distribution')
plt.show()
```
